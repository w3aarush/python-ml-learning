###### Handling Outliers

```python
# boxplot for debtratio
plt.figure()
plt.boxplot(df['DebtRatio'])
plt.title("DebtRatio Boxplot")
plt.ylabel("DebtRatio")
plt.show()
```

```python
df['DebtRatio'].describe()
```

```python
q1 = df['DebtRatio'].quantile(0.25)
q3 = df['DebtRatio'].quantile(0.75)
IQR = q3 - q1
```

```python
IQR
```

```python
lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR
```

```python
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"First Quartile: {q1}")
print(f"Third Quartile: {q3}")
```

```python
df['DebtRatio'] = df['DebtRatio'].clip(lower=lower_bound, upper=upper_bound)
```

```python
df['DebtRatio'].describe()
```

```python
# RevolvingUtilizationOfUnsecuredLines
plt.figure()
plt.boxplot(df['RevolvingUtilizationOfUnsecuredLines'])
plt.title("RevolvingUtilizationOfUnsecuredLines Boxplot")
plt.ylabel("RevolvingUtilizationOfUnsecuredLines")
plt.show()
```

```python
q1 = df['RevolvingUtilizationOfUnsecuredLines'].quantile(0.25)
q3 = df['RevolvingUtilizationOfUnsecuredLines'].quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
```

```python
df['RevolvingUtilizationOfUnsecuredLines'] = df['RevolvingUtilizationOfUnsecuredLines'].clip(lower=lower_bound, upper=upper_bound)
```

###### NumberofTimes90DaysLate

```python
# NumberofTimes90DaysLate
plt.figure()
plt.boxplot(df['NumberOfTimes90DaysLate'])
plt.title("NumberOfTimes90DaysLate Boxplot")
plt.ylabel("NumberOfTimes90DaysLate")
plt.show()
```

```python
q1 = df['NumberOfTimes90DaysLate'].quantile(0.25)
q3 = df['NumberOfTimes90DaysLate'].quantile(0.75)

iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
```

```python
df['NumberOfTimes90DaysLate'].describe()
```

```python
df['NumberOfTimes90DaysLate'].value_counts()
```

```python
df['NumberOfTimes90DaysLate'].isnull().sum()
```

```python
plt.figure()
plt.hist(df['NumberOfTimes90DaysLate'], bins=30)
plt.xlabel('NumberOfTimes90DaysLate')
plt.ylabel("Frequency")
plt.title("NumberOfTimes90DaysLate Distribution (Before Imputation)")
plt.show()
```

```python
# df['NumberOfTimes90DaysLate'] is highly skewed.
# when distribution is heavily concentrated at 0, IQR becomes zero.
# in this case, we perform logical capping
# 10 seems fine, not very less nor high
df['NumberOfTimes90DaysLate'] = df['NumberOfTimes90DaysLate'].clip(upper=10)
```

```python
df['NumberOfTimes90DaysLate'].describe()
```

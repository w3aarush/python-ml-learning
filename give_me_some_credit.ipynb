```python
df['MonthlyIncomeMissing'] = df['MonthlyIncome'].isnull().astype(int)
```

```python
plt.figure()
plt.hist(df['MonthlyIncome'].dropna(), bins=30)
plt.xlabel('Monthly Income')
plt.ylabel("Frequency")
plt.title("Monthly Income Distribution (Before Imputation)")
plt.show()
```

```python
mean_income = df['MonthlyIncome'].mean()
median_income = df['MonthlyIncome'].median()
print(mean_income)
print(median_income)
```

```python
# imputing with median as data is extremely skewed
df['MonthlyIncome'].fillna(median_income, inplace=True)
```

###### Handling Outliers

```python
# boxplot for debtratio
plt.figure()
plt.boxplot(df['DebtRatio'])
plt.title("DebtRatio Boxplot")
plt.ylabel("DebtRatio")
plt.show()
```

```python
df['DebtRatio'].describe()
```

```python
q1 = df['DebtRatio'].quantile(0.25)
q3 = df['DebtRatio'].quantile(0.75)
IQR = q3 - q1
```

```python
IQR
```

```python
lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR
```

```python
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"First Quartile: {q1}")
print(f"Third Quartile: {q3}")
```

```python
df['DebtRatio'] = df['DebtRatio'].clip(lower=lower_bound, upper=upper_bound)
```

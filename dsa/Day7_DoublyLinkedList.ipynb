# Doubly Linked List
<p>It is also a linear data structure. It stores data as well references to the next and previous nodes.<br>Each node contains three components:</p>

* Data
* Reference for Previous node
* Reference for next node

```python
class doublyNode():
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoubleLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_beginning(self, data):
        new_node = doublyNode(data)
        if self.head is not None:
            self.head.prev = new_node
            new_node.next = self.head
        self.head = new_node

    def insert_at_end(self, data):
        new_node = doublyNode(data)
        if self.head is None:
            self.head = new_node
            return

        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        new_node.prev = temp

    def display_forward(self):
        temp = self.head
        while temp:
            print(temp.data, end=" <-> ")
            temp = temp.next
        print("None")

    def display_backward(self):
        temp = self.head
        if temp is None:
            return

        while temp.next:
            temp = temp.next

        while temp:
            print(temp.data, end='<->')
            temp = temp.prev
        print("None")
```

```python
obj1 = DoubleLinkedList()
```

```python
obj1.insert_at_beginning(5)
obj1.insert_at_beginning(40)
obj1.insert_at_end(45)
```

```python
obj1.display_forward()
```

```python
obj1.display_backward()
```

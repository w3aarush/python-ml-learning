* Hash is a data structure that stores key:value pair.
* It provides very fast lookup,insertion, and deletion.
* Python has already a builit in hask like structure called Dictionary.
* Hash function converts a key into index.
* Storage: It is array based.
* Keys are converted to indices.


```python
class hashtable:
    def __init__(self, size=10):
        self.size = size
        self.table = [[] for _ in range(size)]

    def _hash(self, key):
        print(hash(key))
        return hash(key) % self.size

    def insert(self, key, value):
        index = self._hash(key)
        for pair in self.table[index]:
            if pair[0] == key:
                pair[1] = value
                return

        self.table[index].append([key, value])

    def get_value(self, key):
        print(self.table)
        index = self._hash(key)
        print(index)
        for current_item in self.table[index]:
            print("My current item is ", current_item)
            if current_item[0] == key:
                print(current_item[1])

    def display(self):
        for index, bucket in enumerate(self.table):
            print(index, bucket)

```

Create a hashtable to store a list of employee with attributes FirstName, LastName, Salary, Address, Age with size 20

1. Write a insert function to a new employee with the above 5 attributes
2. Write a function to get a employee detail matching a specific Age
3. Write a function to delete an employee from the list with a specific Name
4. Write a function to Display all the employees with salary greater than 10000

<h1>Homework</h1>

```python
class EmployeeDatabase():
    def __init__(self, size=20):
        self.size = size
        self.table = [[] for _ in range(self.size)]

    def _hash(self, key):
        return hash(key) % self.size

    def keygen(self,firstName, lastName):
        return (firstName+lastName).lower()

    def add(self, firstName, lastName, Salary, Address, Age):
        key = self.keygen(firstName,lastName)
        index = self._hash(key)
        for item in self.table[index]:
            if (item[0]+item[1]).lower() == (firstName+lastName).lower():
                print('Already present')
                break

        self.table[index] = [firstName, lastName, Salary, Address, Age]

    def delete_employee(self, firstName, lastName):
        index = self._hash((firstName+lastName).lower())
        self.table[index].clear()

    def get_detail_with_age(self, age):
        fetched_data = []
        for item in self.table:
            if len(item)==0:
                continue
            elif item[4]==age:
                fetched_data.append(item)
        for item in fetched_data:
            print(item)

    def get_salary_greater_than(self):
        fetched_data = []
        for item in self.table:
            if len(item)==0:
                continue
            elif item[2] > 10000:
                fetched_data.append(item)
        for item in fetched_data:
            print(item)

    def display(self):
        for index, bucket in enumerate(self.table):
            print(f"{index} : {bucket}")
```

```python
employee_finance = EmployeeDatabase()
```

```python
employee_finance.add('Person', '1', 50000, "Patna, Bihar", 24)
employee_finance.add('Person', '2', 12000, "Location 2", 25)
employee_finance.add('Person', '3', 10000, "Location 3", 26)
employee_finance.add('Person', '4', 6000, 'Location 4', 24)
```

<h1>Stacks & Queue Homework</h1>

<p>A cinema has one ticket counter.<br>People arrive, stand in a queue, buy tickets, and leave.</p>

**RULES**

* People join the queue -> enqueue(person)
* Counter serves the first person -> dequeue()
* Show next person -> peek()
* If queue is empty -> show "No Customers"

```python
import queue
```

```python
ticket_counter = queue.Queue()
```

```python
def enqueue(person):
    ticket_counter.put(person)
    print(f"{person} joined the queue.")
```

```python
def dequeue():
    if not ticket_counter.empty():
        person_with_ticket = ticket_counter.get()
        print(f"{person_with_ticket} got the ticket.")
    else:
        print("No Customers")
```

```python
def peek():
    if not ticket_counter.empty():
        print(f"{ticket_counter.queue[0]} is next in line.")
    else:
        print("No Customers")
```

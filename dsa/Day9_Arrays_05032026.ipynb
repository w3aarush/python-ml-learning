Array is a data structure that stores a collection of elements of the same type in a contiguous memory location. Elements are stored sequentially.
* address of element = base_address+(index * sizeofelement)
* base_address = 1000 and each element element = 4 bytes
* element at 0 index has address 1000
* element at 1 index will have 1000 + 1 * 4 = 1004 as address

Create a dynamic 2d array with 3 rows and 4 columns and each value should be multiplied by 3.


```python
row = 3
col = 4
lst = []
mul = 3
for i in range(row):
    con_list = []
    for j in range(col):
        con_list.append(mul)
        mul += 3
    lst.append(con_list)
print(lst)
```

```python
def transpose(matrix_list):
    matrix_row = len(matrix_list)
    matrix_col = len(matrix_list[0])
    transposed = []
    for i in range(matrix_col):
        transposed_row = []
        for item in matrix_list:
            transposed_row.append(item[i])
        transposed.append(transposed_row)
    return transposed
```

```python
for item in lst:
    print(item)
```

```python
transposed_lst = transpose(lst)
for item in transposed_lst:
    print(item)
```

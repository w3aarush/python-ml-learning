# Singly Cicular Linked List
* Each node has a data it's last node connect to the first node.

```python
class SingleCircularNode:
    def __init__(self, data):
        self.data = data
        self.next = None

class SingleCircularLinkedList:
    def __init__(self):
        self.head = None

    def insertNode(self, data):
        new_node = SingleCircularNode(data)
        if self.head is None:
            self.head = new_node
            new_node.next = self.head
            return

        current = self.head
        while current.next != self.head:
            current = current.next

        current.next = new_node
        new_node.next = self.head

    def display(self):
        if self.head is None:
            return
        current = self.head
        while current:
            print(f"{current.data}")
            current = current.next
            if current == self.head:
                return
```

```python
obj = SingleCircularLinkedList()
```

```python
obj.insertNode(5)
obj.insertNode(7)
obj.insertNode(9)
obj.display()
```

# Homework
<p>Using Doubly circular LinkedList, creat nodes in such a way that each Node consist below attributes.</p>


* song_name
* artist
* duration
* prev
* next

<p>Write a function to display next music information.</p>
<p>Write a function to display previous music information.</p>
<p>Write insert to add new nodes</p>

```python
class DoublyCircularNode:
    def __init__(self, song_name, artist, duration):
        self.song_name = song_name
        self.artist = artist
        self.duration = duration

class DoublyCircularLinkedList:
    def __init__(self):
        self.head = None

    def insertNode(self, song_name, artist, duration):
        new_node = DoublyCircularNode(song_name, artist, duration)
        if self.head is None:
            self.head = new_node
            new_node.next = self.head
            new_node.prev = self.head
            return

        current = self.head
        while current.next != self.head:
            current = current.next

        current.next = new_node
        new_node.prev = current
        new_node.next = self.head
        self.head.prev = new_node

    def display(self):
        if self.head is None:
            return
        current = self.head
        while current.next != self.head:
            print(f"Song Name: {current.song_name}\nArtist: {current.artist}\nDuration: {current.duration}\n")
            current = current.next
```

```python
obj = DoublyCircularLinkedList()
```

```python
obj.insertNode("Jug Jug Jeeve", "Unknown", "02:30")
```

```python
obj.display()
```

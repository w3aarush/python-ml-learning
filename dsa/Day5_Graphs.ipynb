Types of graph:
> undirected graphs - in which nodes are connected both ways
> directed graph -

```python
from collections import defaultdict
```

```python
class undirected_graph:
    def __init__(self):
        self.graph = defaultdict(list)

    def add_edge(self, v1, v2):
        self.graph[v1].append(v2)
        self.graph[v2].append(v1)
    def display(self):
        for vertex, neighbours in self.graph.items():
            print(vertex, '-->', neighbours)
```

```python
g1 = undirected_graph()
```

```python
g1.add_edge('node1','node2')
```

```python
g1.display()
```

```python
g1.add_edge('node2','node3')
```

```python
g1.display()
```

## Directed Graph
<p>Directed grah is the one way grah, where one node can go towards or other but the other node can't travel back to it.</p>
<p>E.g., Following someone on linkedin</p>

```python
from collections import defaultdict
```

```python
class Dgraph:
    def __init__(self):
        self.graph = defaultdict(list)

    def add_edge(self, v1, v2):
        self.graph[v1].append(v2)

    def display(self):
        for vertex, neighbours in self.graph.items():
            print(vertex, "follows", neighbours)
```

```python
dgraph = Dgraph()
```

```python
dgraph.add_edge('node1','node2')
```

```python
dgraph.display()
```

Weighted Directed graph - every edge has weight associated to it and weight can be in the form of time or cost.
e.g. - google maps showing time and distance b/w two destinations.

```python
class WeightedDirectedGraph:
    def __init__(self):
        self.graph = {}

    def add_vertex(self, vertex):
        if vertex not in self.graph:
            self.graph[vertex] = []

    def add_edge(self, source, destination, weight):
        if source not in self.graph:
            self.add_vertex(source)
        if destination not in self.graph:
            self.add_vertex(destination)

        self.graph[source].append((destination, weight))

    def display(self):
        for vertex in self.graph:
            print(f"{vertex} -> {self.graph[vertex]}")


# Example usage
g = WeightedDirectedGraph()
g.add_edge("A", "B", 5)
g.add_edge("A", "C", 3)
g.add_edge("B", "D", 2)

g.display()
```

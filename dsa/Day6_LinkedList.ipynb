* Linked list a linear data structure where we store reference to a node.
* linked list consists of various nodes.
* each node consist of two things: data and reference to next node.
* Types of linked list:
  * Singly Linked List
  * Doubly Linked List
  * Circular Linked List



```python
class Node():
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList(): # nodes only point to the next node, can't go back to the previous node
    def __init__(self):
        self.head = None

    def insertAtBeginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def insertAtEnd(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next

        temp.next = new_node

    def display(self):
        temp = self.head
        while temp:
            print(temp.data,"Next node",temp.next)

        print("None")

```

Perform below actions

```python
class Node():
    def __init__(self, task_id, task_name, priority):
        self.task_id = task_id
        self.task_name = task_name
        self.priority = priority
        self.next = None

class LinkedList():
    def __init__(self):
        self.head = None

    def add_node_beginning(self, task_id, task_name, priority):
        new_node = Node(task_id, task_name, priority)
        new_node.next = self.head
        self.head = new_node

    def add_node_end(self, task_id, task_name, priority):
        new_node = Node(task_id, task_name, priority)
        if self.head is None:
            self.head = new_node
        else:
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_node

    def add_priority_ascending(self, task_id, task_name, priority):
        new_node = Node(task_id, task_name, priority)

        if self.head is None or new_node.priority < self.head.priority:
            new_node.next = self.head
            self.head = new_node
        else:
            temp = self.head
            while temp.next and temp.next.priority <= new_node.priority:
                temp = temp.next
            new_node.next = temp.next
            temp.next = new_node

    def display(self):
        current = self.head
        nodes = []
        while current:
            nodes.append(f"({current.task_id}, {current.task_name}, {current.priority})")
            current = current.next
        print(" --> ".join(nodes) + " --> None")
```

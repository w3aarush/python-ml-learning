## 3. ***Data Wrangling***

### Data Wrangling Code

````python
df[(df['price'] == 0) & (df['availability_365'] == 0)].head()
````

````python
df.drop(df[df['host_id'] == 13709292].index,axis = 0, inplace = True)
````

````python
df.describe()
````

````python
# boxplot for df['price']
plt.figure(figsize=(10, 10))
plt.boxplot(df['price'])
plt.show()
````

````python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 # Ensure lower bound is not less than 0
upper_bound = Q3 + 1.5 * IQR

df['price'].clip(lower=lower_bound, upper=upper_bound,inplace=True)
````

````python
plt.boxplot(df['price'])
plt.show()
````

````python
plt.boxplot(df['minimum_nights'])
plt.show()
````
